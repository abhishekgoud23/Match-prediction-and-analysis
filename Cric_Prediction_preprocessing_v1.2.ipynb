```python
df = df.drop(columns=['name','description','start_date','end_date','venue_id','points','highlights','home_playx1','away_playx1','umpire1','umpire2','tv_umpire','referee','reserve_umpire'])
```

```python
df
```

```python
df.dropna(inplace=True)
df
```

```python
df = df[~df.result.str.contains("abandoned")]
df
```

```python
df = df[~df.result.str.contains("No result")]
df
```

```python
Q1 = df[['home_score', 'away_score']].quantile(0.25)
Q3 = df[['home_score', 'away_score']].quantile(0.75)

# Calculate the IQR for each column
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Replace the outliers with NaN values
df[['home_score', 'away_score']] = df[['home_score', 'away_score']].mask((df[['home_score', 'away_score']] < lower_bound) | (df[['home_score', 'away_score']] > upper_bound), np.nan)
```

```python

# Drop rows with NaN values
df.dropna(inplace=True)

df
```

```python
df = df.rename(columns={'short_name': 'match_between', 'venue_name': 'venue', 'pom': 'player_of_the_match'})
df
```

```python
# Extracting part of the day from 'match_days' column using regex

word_before = '-'
word_after = 'match'
df['part_of_the_day'] = df['match_days'].str.extract(fr"{word_before}\s+(?P<word>\w+)\s+{word_after}")
df['part_of_the_day'].fillna(value = 'day/night', inplace = True)
df
```

```python
#Drop the 'match_days' column

df.drop('match_days', axis = 1, inplace = True)
df
```

```python
def checkChar(target_string, char, append_string):
    if char in target_string:
        return target_string
    else:
        return target_string + append_string
df['1st_inning_score'] = df['1st_inning_score'].apply(checkChar, args=('/', '/10'))
df['2nd_inning_score'] = df['2nd_inning_score'].apply(checkChar, args=('/', '/10'))
df['home_score'] = df['home_score'].apply(checkChar, args=('/', '/10'))
df['away_score'] = df['away_score'].apply(checkChar, args=('/', '/10'))
df
```

```python
df = df.replace('KXIP' , 'PBKS' , regex = True)
```

```python
df.to_csv('cleaned_Cric_summary.csv')
```
