```python
import pandas as pd
data = pd.read_csv('cleaned_Cric_summary.csv')
```

```python
data = data.dropna()       # drop any rows with missing values
data
```

```python
# extract score and convert to integer
data['1st_inning_runs'] = data['1st_inning_score'].str.split('/').str[0].astype(int)
data['1st_inning_wkts'] = data['1st_inning_score'].str.split('/').str[1].astype(int)

data['2nd_inning_runs'] = data['2nd_inning_score'].str.split('/').str[0].astype(int)
data['2nd_inning_wkts'] = data['2nd_inning_score'].str.split('/').str[1].astype(int)

data
```

```python
data.drop(['Unnamed: 0','id','match_between','1st_inning_score','2nd_inning_score','home_score','away_score','result','part_of_the_day'], axis=1)
```

```python
data['is_home_winner'] = (data['home_team'] == data['winner']) * 1
data
```

```python
from sklearn.preprocessing import LabelEncoder         # import label encoder from sklearn for encoding categorical data, and select the columns to encode

labelEnc = LabelEncoder()
colsForEncoding = ['home_team', 'away_team', 'toss_won', 'decision', 'venue', 'home_captain', 'away_captain', 'player_of_the_match','home_key_batsman','home_key_bowler','away_key_batsman','away_key_bowler']
for i in colsForEncoding:                              # iterate through the columns and encode the categorical data using label encoding
    data.loc[:, i] = labelEnc.fit_transform(data.loc[:, i])
```

```python
data
```

```python
from pyspark.sql import SparkSession                   # import spark and create a spark session, then convert the pandas dataframe to spark dataframe
spark = SparkSession.builder.appName("CricPrediction").getOrCreate()
df = spark.createDataFrame(data)
```

```python
df = df.drop('Unnamed: 0','id','1st_inning_score','2nd_inning_score','home_score','away_score','result','part_of_the_day')
df.show()
```

```python
from pyspark.sql.functions import col, avg

# Run rate of each team
df = df.withColumn('home_run_rate', col('home_runs')/col('home_overs'))
df = df.withColumn('away_run_rate', col('away_runs')/col('away_overs'))

# Economy rate of each team
df = df.withColumn('home_economy_rate', col('away_runs')/col('home_overs'))
df = df.withColumn('away_economy_rate', col('home_runs')/col('away_overs'))


# The difference between the scores of each team in the 1st inning and 2nd inning
df = df.withColumn('score_difference', col('home_runs') - col('away_runs'))

# The difference between the number of wickets taken by each team in the 1st inning and 2nd inning
df = df.withColumn('wicket_difference', col('home_wickets') - col('away_wickets'))


df.show()
```

```python
from pyspark.ml.feature import VectorAssembler, StringIndexer

# Define the feature columns
featuresCol = ['home_team','away_team','toss_won','decision', 'venue','home_captain','away_captain','player_of_the_match','home_overs','home_runs','home_wickets','home_boundaries','away_overs','away_runs','away_wickets','away_boundaries','1st_inning_runs','1st_inning_wkts','2nd_inning_runs','2nd_inning_wkts','home_run_rate','away_run_rate','home_economy_rate','away_economy_rate','score_difference','wicket_difference']

# Assemble the features into a single vector column
#Creating a VectorAssembler object assemble that will assemble the feature columns defined above into a single vector column called 'features'.

assemble = VectorAssembler(inputCols=featuresCol, outputCol='features')

df = assemble.transform(df)
indexer = StringIndexer(inputCol="is_home_winner", outputCol="label")
df = indexer.fit(df).transform(df)

label_col = 'label'
df = df.select(['features', label_col])
```

```python
train, test = df.randomSplit([0.7, 0.3], seed=123)
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

rf = RandomForestClassifier(numTrees=100, maxDepth=4, labelCol=label_col, seed=123)
rf_model = rf.fit(train)
```

```python
rf_predictions = rf_model.transform(test)
# Compute evaluation metrics such as accuracy, precision, recall, F1-score, etc.
rf_evaluator = MulticlassClassificationEvaluator(labelCol=label_col, metricName="accuracy")
rf_accuracy = rf_evaluator.evaluate(rf_predictions)
rf_precision = rf_evaluator.evaluate(rf_predictions, {rf_evaluator.metricName: "weightedPrecision"})
rf_recall = rf_evaluator.evaluate(rf_predictions, {rf_evaluator.metricName: "weightedRecall"})
rf_f1 = rf_evaluator.evaluate(rf_predictions, {rf_evaluator.metricName: "f1"})

# Print the evaluation metrics
print("Accuracy: %g" % rf_accuracy)
print("Precision: %g" % rf_precision)
print("Recall: %g" % rf_recall)
print("F1-score: %g" % rf_f1)
```

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

# Create a ParamGridBuilder with hyperparameters to try
param_grid = (ParamGridBuilder()
             .addGrid(rf.numTrees, [50, 100, 150])
             .addGrid(rf.maxDepth, [2, 4, 6])
             .build())

# Create a CrossValidator with the Random Forest Classifier and ParamGridBuilder
cv = CrossValidator(estimator=rf, estimatorParamMaps=param_grid, evaluator=rf_evaluator, numFolds=5, seed=123)

# Fit the CrossValidator to the training data
cv_model = cv.fit(train)

# Get the best model from the CrossValidator
best_model = cv_model.bestModel

predictions = best_model.transform(test)
# Compute evaluation metrics such as accuracy, precision, recall, F1-score, etc.
rft_evaluator = MulticlassClassificationEvaluator(labelCol=label_col, metricName="accuracy")
rft_accuracy = rft_evaluator.evaluate(predictions)
rft_precision = rft_evaluator.evaluate(predictions, {rft_evaluator.metricName: "weightedPrecision"})
rft_recall = rft_evaluator.evaluate(predictions, {rft_evaluator.metricName: "weightedRecall"})
rft_f1 = rft_evaluator.evaluate(predictions, {rft_evaluator.metricName: "f1"})

# Print the evaluation metrics
print("Accuracy: %g" % rft_accuracy)
print("Precision: %g" % rft_precision)
print("Recall: %g" % rft_recall)
print("F1-score: %g" % rft_f1)
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Split the data into training and test sets
train, test = df.randomSplit([0.7, 0.3], seed=123)

# Create a LogisticRegression object and fit the model to the training data
lr = LogisticRegression(labelCol=label_col, featuresCol="features", maxIter=10, regParam=0.01)
lr_model = lr.fit(train)

# Use the model to make predictions on the test data
lr_predictions = lr_model.transform(test)

# Compute evaluation metrics such as accuracy, precision, recall, F1-score, etc.
lr_evaluator = MulticlassClassificationEvaluator(labelCol=label_col, metricName="accuracy")
lr_accuracy = lr_evaluator.evaluate(lr_predictions)
lr_precision = lr_evaluator.evaluate(lr_predictions, {lr_evaluator.metricName: "weightedPrecision"})
lr_recall = lr_evaluator.evaluate(lr_predictions, {lr_evaluator.metricName: "weightedRecall"})
lr_f1 = lr_evaluator.evaluate(lr_predictions, {lr_evaluator.metricName: "f1"})

# Print the evaluation metrics
print("Accuracy: %g" % lr_accuracy)
print("Precision: %g" % lr_precision)
print("Recall: %g" % lr_recall)
print("F1-score: %g" % lr_f1)
```

```python
from pyspark.ml.classification import LinearSVC
from pyspark.ml.evaluation import MulticlassClassificationEvaluator


# Split the data into training and test sets
train, test = df.randomSplit([0.7, 0.3], seed=123)

svm = LinearSVC(labelCol=label_col, featuresCol="features", maxIter=10, regParam=0.1)

# Define the evaluator
svm_evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='label', metricName='accuracy')

# Define the parameter grid
paramGrid = (ParamGridBuilder()
             .addGrid(svm.regParam, [0.1, 0.01])
             .addGrid(svm.maxIter, [10, 20])
             .build())

# Define the cross validator
cv = CrossValidator(estimator=svm, estimatorParamMaps=paramGrid, evaluator=svm_evaluator, numFolds=5)

# Fit the cross validator to the train data
svm_model = cv.fit(train)

# Make predictions on the test data
svm_predictions = svm_model.transform(test)
svm_evaluator = MulticlassClassificationEvaluator(labelCol=label_col, metricName="accuracy")


# Evaluate the model performance using the evaluation metric
gbt_accuracy = svm_evaluator.evaluate(svm_predictions)
gbt_precision = svm_evaluator.evaluate(svm_predictions, {svm_evaluator.metricName: "weightedPrecision"})
gbt_recall = svm_evaluator.evaluate(svm_predictions, {svm_evaluator.metricName: "weightedRecall"})
gbt_f1 = svm_evaluator.evaluate(svm_predictions, {svm_evaluator.metricName: "f1"})

# Print the evaluation metrics
print("Accuracy: %g" % gbt_accuracy)
print("Precision: %g" % gbt_precision)
print("Recall: %g" % gbt_recall)
print("F1-score: %g" % gbt_f1)
```

```python
from pyspark.ml.classification import GBTClassifier

train, test = df.randomSplit([0.7, 0.3], seed=123)

# Initialize the Gradient Boosting classifier
gbt = GBTClassifier(labelCol=label_col, featuresCol="features", maxIter=10)

# Define the hyperparameter grid to search
param_grid = ParamGridBuilder() \
    .addGrid(gbt.maxDepth, [2, 4, 6]) \
    .addGrid(gbt.stepSize, [0.05, 0.1, 0.2]) \
    .build()

# Define the evaluation metric
gbt_evaluator = MulticlassClassificationEvaluator(labelCol=label_col, metricName="accuracy")

# Define the cross-validation object
gbt_cv = CrossValidator(estimator=gbt, estimatorParamMaps=param_grid, evaluator=gbt_evaluator, numFolds=5)

# Fit the model using the cross-validation object
gbt_model = gbt_cv.fit(train)

# Make predictions on the test set
gbt_predictions = gbt_model.transform(test)

# Evaluate the model performance using the evaluation metric
gbt_accuracy = gbt_evaluator.evaluate(gbt_predictions)
gbt_precision = gbt_evaluator.evaluate(gbt_predictions, {gbt_evaluator.metricName: "weightedPrecision"})
gbt_recall = gbt_evaluator.evaluate(gbt_predictions, {gbt_evaluator.metricName: "weightedRecall"})
gbt_f1 = gbt_evaluator.evaluate(gbt_predictions, {gbt_evaluator.metricName: "f1"})

# Print the evaluation metrics
print("Accuracy: %g" % gbt_accuracy)
print("Precision: %g" % gbt_precision)
print("Recall: %g" % gbt_recall)
print("F1-score: %g" % gbt_f1)
```
